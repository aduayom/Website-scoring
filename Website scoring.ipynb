## Analyse et Optimisation de la Vitesse de Chargement avec PageSpeed Insights

La page PageSpeed est en réalité un service en ligne fourni par Google, connu sous le nom de "PageSpeed Insights" (auparavant "Google PageSpeed"). Son objectif principal est d'analyser et d'améliorer la vitesse de chargement d'une page web. PageSpeed Insights propose les fonctionnalités suivantes :

- **Analyse de la Vitesse de Chargement** : L'outil évalue la performance de chargement d'une page en prenant en compte divers indicateurs, tels que le temps nécessaire pour que la page soit prête à l'utilisation, la rapidité d'affichage du contenu principal, et d'autres métriques liées à la performance.

- **Évaluation des Performances** : PageSpeed Insights attribue une note de performance à la page, notée de 0 à 100, pour évaluer la rapidité de chargement. Une note plus élevée indique une meilleure performance.

- **Recommandations d'Optimisation** : Le service fournit des recommandations détaillées pour améliorer la vitesse de chargement. Ces conseils peuvent inclure la compression d'images, la mise en cache de contenu, la réduction du code JavaScript et CSS, l'élimination de redirections inutiles, l'amélioration de l'expérience mobile, et d'autres optimisations.

- **Analyse de la Version Mobile** : PageSpeed Insights propose une analyse spécifique de la version mobile de la page, mettant en lumière les problèmes de performance liés aux appareils mobiles.

- **Prise en Charge des Technologies Web Modernes** : L'outil tient compte des meilleures pratiques et des nouvelles technologies web pour aider à optimiser la vitesse de chargement et l'expérience utilisateur.

- **Intégration avec Lighthouse** : PageSpeed Insights s'appuie sur Lighthouse, un outil open-source qui offre une analyse avancée de la performance web, de l'accessibilité, de l'expérience utilisateur, et plus encore.

En utilisant PageSpeed Insights, les propriétaires de sites web et les développeurs peuvent évaluer et améliorer la vitesse de chargement de leurs pages, ce qui peut avoir un impact significatif sur l'expérience des utilisateurs, le référencement et la conversion des visiteurs en clients. ## Récupération des différents indicateurs
Dans le cadre de la constitution de notre base de données, nous allons collecter des indicateurs essentiels qui se divisent en deux catégories distinctes. La première vague d'indicateurs comprend :

- **Largest Contentful Paint (LCP)** : Cet indicateur mesure le temps nécessaire pour que le plus grand élément visible à l'écran d'une page web se charge complètement. Il est essentiel pour évaluer la rapidité de chargement d'une page.

- **First Input Delay (FID)** : Le FID mesure la réactivité d'une page web en enregistrant le délai entre l'interaction de l'utilisateur (par exemple, un clic) et la réponse du site. Un FID faible est crucial pour offrir une expérience utilisateur fluide.

- **Cumulative Layout Shift (CLS)** : Le CLS évalue la stabilité de la mise en page d'une page web. Il prend en compte les changements inattendus dans la disposition des éléments, ce qui peut perturber l'expérience de l'utilisateur.

- **First Contentful Paint (FCP)** : Le FCP indique le moment où le premier élément visible à l'écran commence à se charger. C'est un indicateur clé pour évaluer la vitesse de chargement perceptible par l'utilisateur.

- **Interaction to Next Paint (INP)** : L'INP mesure le temps écoulé entre une interaction de l'utilisateur et le début du chargement du prochain élément à l'écran. Cela influence directement l'expérience de l'utilisateur.

- **Time to First Byte (TTFB)** : Le TTFB est le temps nécessaire pour que le serveur web réponde à la demande initiale, ce qui influe sur le temps de chargement global d'une page.

Ces indicateurs joueront un rôle essentiel dans l'évaluation des performances des sites web que nous analyserons, contribuant ainsi à la constitution de notre base de données.

![Logo Markdown](https://imageshack.com/i/pohOIo8ep)


```python
def First_kpi(soup,Dict):
    # Trouver la div principale de classe "OyzGL"
    div_principale = soup.find('div', class_='OyzgL')
    # Si la div principale est trouvée
    if div_principale:
        # Trouver toutes les divs de classe "PZeSbe" à l'intérieur de la div principale
        divs_secondaire = div_principale.find_all('div', class_='PZeSbe')

        # Parcourir toutes les divs de classe "PZeSbe"
        for span in divs_secondaire:
            # Trouver le span avec la classe "h1llh" dans la div "PZeSbe" actuelle
            span_h1llh = span.find('span', class_='h1llh')
            Nom_kpi = span_h1llh.text if span_h1llh else "Span h1llh introuvable"

            # Trouver le span avec la classe "Ykn2A" dans la div "PZeSbe" actuelle
            span_Ykn2A = span.find('span', class_='Ykn2A')
            valeur_kpi = span_Ykn2A.text if span_Ykn2A else "Span Ykn2A introuvable"

            # Affichez le contenu des spans pour la div "PZeSbe" actuelle
            #print("Contenu de span h1llh:", Nom_kpi)
            #print("Contenu de span Ykn2A:", valeur_kpi)
            Dict[Nom_kpi] = valeur_kpi

    else:
        print("Div principale introuvable")
    return Dict
    
```

La seconde vague d'indicateurs porte sur l'analyse des problèmes de performances, accessbilité, bonnes pratiques, et seo que nous recupérons à travers la fonction second_kpi

### Seconde Vague d'Indicateurs

La seconde vague d'indicateurs se penche sur l'analyse des problèmes liés à la performance, à l'accessibilité, aux bonnes pratiques et au référencement (SEO). Ces données sont recueillies grâce à la fonction second_kpi. Les indicateurs de la seconde vague incluent :

- **Performances** : Évalue la qualité globale de la performance d'un site web.

- **Accessibilité** : Analyse l'accessibilité d'un site pour tous les utilisateurs, y compris ceux ayant des besoins spéciaux.

- **Bonnes pratiques** : Évalue le respect des meilleures pratiques en matière de développement web.

- **SEO (Search Engine Optimization)** : Évalue l'optimisation du site pour les moteurs de recherche, ce qui influe sur son classement dans les résultats de recherche.

Ces indicateurs contribuent à évaluer la qualité globale des sites web que nous analysons et nous aident à constituer notre base de données.



![Logo Markdown](https://imagizer.imageshack.com/img922/381/jyE38s.png)


```python
def Second_Kpi(soup,Dict):
    div_principale = soup.find('div', class_='lh-scores-header')
    # Si la div principale est trouvée
    if div_principale:
        # Trouver toutes les divs de classe "PZeSbe" à l'intérieur de la div principale
        All_A = div_principale.find_all('a', class_='lh-gauge__wrapper lh-gauge__wrapper--average')

        # Parcourir toutes les divs de classe "PZeSbe"
        for div in All_A:
            # Trouver le span avec la classe "h1llh" dans la div "PZeSbe" actuelle
            label = div.find('div', class_='lh-gauge__label')
            label_text = label.text if label else "label introuvable"

            # Trouver le span avec la classe "Ykn2A" dans la div "PZeSbe" actuelle
            valeur = div.find('div', class_='lh-gauge__percentage')
            valeur_text = valeur.text if valeur else "valeur introuvable"

            # Affichez le contenu des spans pour la div "PZeSbe" actuelle
            #print("label:", label_text)
            #print("valeur:", valeur_text)
            Dict[label_text] = valeur_text
    else:
        print("Div principale introuvable")
    return Dict
    
```

```python
def Extract(Link):
    # Créer une instance du navigateur
    driver = webdriver.Chrome()

    # Ouvrir le site web
    driver.get("https://pagespeed.web.dev/?hl=fr")  # URL du site de nos performances

    # Trouver l'élément de la barre de recherche en utilisant son nom de class
    barre_de_recherche = driver.find_element(By.CLASS_NAME, "VfPpkd-fmcmS-wGMbrd")

    # Saisir le texte dans la barre de recherche
    texte_a_rechercher = Link
    barre_de_recherche.send_keys(texte_a_rechercher)

    # Soumettre la recherche en appuyant sur la touche Entrée
    barre_de_recherche.send_keys(Keys.RETURN)
    wait = WebDriverWait(driver, 60)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, "lh-scores-header")))
    Dict={}
    Dict["Liens"] = Link
    ######################### Récupération de la première partie des données #########################
    # Utiliser Beautiful Soup pour analyser la page
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    Dict=First_kpi(soup,Dict)
    ######################### Récupération de la seconde partie des données #########################
    Dict=Second_Kpi(soup,Dict)    
    # Fermer le navigateur
    driver.quit()
    return Dict
```

# Nettoyage des Données : Transformation des Valeurs Numériques

Nous entamons maintenant la phase cruciale du nettoyage de notre ensemble de données. Au cours de cette étape, nous avons observé que certaines valeurs numériques sont accompagnées d'unités de mesure. Pour simplifier notre manipulation des données, notre objectif est de convertir ces colonnes en types entiers (int).

## Rationalisation de la Donnée

Une observation importante que nous avons faite est que certaines valeurs numériques dans notre jeu de données sont exprimées avec des unités de mesure, telles que "s" pour les secondes ou "ms" pour les millisecondes. Cela peut compliquer l'analyse ultérieure.

## Transformation en Types Entiers

Pour résoudre ce problème, nous allons procéder à la transformation de ces colonnes en types entiers (int), en éliminant les unités de mesure inutiles. Cette opération nous permettra de disposer de données numériques pures, prêtes à être manipulées de manière plus efficace.

Cette étape de nettoyage est essentielle pour préparer nos données en vue d'analyses plus avancées, et elle illustre l'importance de rationaliser les données pour garantir leur qualité et leur cohérence.


```python
# Définissez une fonction pour extraire la partie numérique
def extraire_numerique(valeur):
    if isinstance(valeur, str):  # Vérifiez si la valeur est une chaîne de caractères
        try:
            # Essayez de convertir la valeur en nombre
            return float(((valeur.replace(',', '.')).replace('s', '')).replace('m', ''))
        except ValueError:
            # En cas d'erreur de conversion, renvoyez une valeur manquante
            return None
    else:
        return valeur

# Appliquez la fonction à chaque colonne de la DataFrame
for colonne in Df.columns:
    if colonne != 'Liens' :
        Df[colonne] = Df[colonne].apply(lambda x: extraire_numerique(x))

# Affichez la DataFrame transformée
Df.head()
```
